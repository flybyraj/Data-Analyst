#### Dictionaries

Dictionaries are unordered collection of items . They store data in Key-Value pairs. Keys must be unique and immutable (e.g strings , numbers or tuples), while values can be of any type.

```python
## Creating Dictionaries

empty_dict={}
print(type(empty_dict))
```

```python
empty_dict=dict()
empty_dict
```

```python
student={"name":"Raj","age":22,"grade":22}
print(student)
print(type(student))
```

```python
## Error Single Key is always used 

student={"name":"Raj","age":22,"name":22}
print(student)
```

```python
## Accessing Dictionary Elements

student={"name":"Raj","age":22,"grade":'A'}
print(student)

print(student["grade"])
print(student['age'])

## Accessing using get() method

print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name',"Not Available"))
```

```python
## Modifying Dictionary Elements
## Dictionary are mutable , so you can add , update and delete Elements
```

```python
student["age"]=90  ## Updated Value for the Key
print(student)
student["address"]="India" ## added a new key value
print(student)
```

```python
del student["grade"] ## delete key and value pair

print(student)
```

```python
## Dictionary methods

keys=student.keys()  ## get all the Keys
print(keys)
values=student.values() ## get all Values
print(values)

items=student.items() ## get all key value Pairs
print(items)
```

```python
##  shallow copy

student_copy=student
print(student)
print(student_copy)
```

```python
student["name"]="Sharma"
print(student)
print(student_copy)
```

```python
student_copy1=student.copy()  ## shallow copy
print(student_copy1)
print(student)
```

```python
student["name"]="Sharma222"
print(student)
print(student_copy1)
```

```python
## Iterating over Dictionaries
## You can use loops to iterate over dictionaries , key ,values, or items(both)


## Iterating Over Keys
for keys in student.keys():
    print(keys)
```

```python
## Iterating over values
for values in student.values():
    print(values)
```

```python
## Iterating over Key:Value Pairs
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
## Nested Dictionaries
## Dictionary inside Dictionary

student={
    "student1":{"name":"raj","age":22},
    "student2":{"name":"sharma","age":20}
    }
print(student)
```

```python
## Acccess nested Dictionary Elements

print(student["student2"]["name"])
print(student["student2"]["age"])
```

```python
## Iterating over Nested Dictionaries

for student_id,student_info in student.items():
    print(f"{student_id}:{student_info}")
    for key,values in student_info.items():
        print(f"{key}:{values}")
```

```python
## Dictionary Comphrehension

squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## Conditional DIctionary Comprehension

evens={x:x**2 for x in range(10) if x%2==0}
print(evens)
```

```python
## Practical Examples 

## Use a dictionary to count frequency of elements in list

numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}
    
for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
## Merge 2 Dictionaries into One

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}
merged_dict={**dict1,**dict2}
print(merged_dict)
```
