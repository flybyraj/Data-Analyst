####Sets

Sets are a buitl-in data type inPython used to store collection of unique items. They are unordereed meaning the elements do not follow a specific order , and they do not allow duplicate elements.
Sets are useful for membership teste , eliminating duplicate entries , and performing mathematical set operations like union , intersection , difference , and symmetric difference.

```python
## Create a set

my_set={1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set=set()
print(type(my_empty_set))
```

```python
my_set=set([1,2,3,4,5,6])
print(type(my_empty_set))
print(my_set)
```

```python
my_empty_set=set([1,2,3,4,5,6,6,5])
print(my_empty_set)
```

```python
## Basic Set Operations
## Adding and Removing Elements


my_set.add(7)
print(my_set)
my_set.add(7)
print(my_set)
```

```python
## Removing Elements from a Set

my_set.remove(3)
print(my_set)
```

```python
my_set.remove(10)
```

```python
my_set.discard(11)
print(my_set)
```

```python
## Pop Method
removed_element=my_set.pop()
print(removed_element)
print(my_set)
```

```python
## Clear all the Elements
my_set.clear()
print(my_set)
```

```python
## Set Membership Test
my_set={1,2,3,4,5}
print(3 in my_set)
print(10 in my_set)
```

```python
## Mathematical Operations

set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

### Union
union_set=set1.union(set2)
print(union_set)

## Intersection
intersection_set=set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2)
print(set1)
```

```python
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

## Difference 
print(set1.difference(set2))

## difference_update works same
```

```python
set2.difference(set1)
```

```python
## Symmetric Difference

set1.symmetric_difference(set2)
```

```python
## Sets Methods

set1={1,2,3,4,5}
set2={3,4,5}

## issubset
print(set1.issubset(set2))

print(set1.issuperset(set2))
```

```python
lt=[1,2,2,3,4,4,5]

set(lt)
```

```python
### Counting Unique words in Text

text="In this tutorial we are discussing about sets"
words=text.split()

## Convert list of words to set to get unique words

unique_words=set(words)
print(unique_words)
print(len(unique_words))
```
