#### Functions in Python

Definition:

A function is a block of code that performs a specific task . FUnctions help in organizing code , reusing code and improving readability.

```python
#Syntax

def function_name(parameter):
    """Docstring"""
    # Function Body
    return expression
```

```python
## Why functions?

num=24
if num%2==0:
    print("Even")
else:
    print("Odd")
```

```python
def even_or_odd(num):
    """This function finds even or odd"""
    if num%2==0:
        print("Even")
    else:
        print("Odd")
```

```python
## Call this Function
even_or_odd(45)
```

```python
## Function with Multiple Parameters

def add(a,b):
    return a+b

result=add(2,4)
print(result)
```

```python
## Default Parameter

def greet(name):
    print(f"hello {name}")

greet("Raj")
```

```python
def greet(name="Guest"):
    print(f"hello {name}")

greet()
```

```python
## Variable Lenght Arguments
## Positional and Keywords argumenta

def print_numbers(*args):  ## Positional Arguments
    for numbers in args:
        print(numbers)
    
```

```python
print_numbers(1,2,3,4,5,6,7,"Raj Sharma")
```

```python
## Keywords Argument

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="Raj",age="32",country="India")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional Argument : {val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,5,"Raj",name="Raj",age=22,country="India")
```
