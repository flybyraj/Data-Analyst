### Data Manipulation and Analysis with Pandas

Data amnipuation and analysis are key tasks in any data analysis project. Pandas provides a wide range of functions for data manipulation and analysis , making it easier to clean , transform and extract insights from data. 

```python
import pandas as pd
```

```python
df=pd.read_csv('data.csv')

## fetch first five rows
df.head(5)
```

```python
## Fetch last five rows
df.tail(5)
```

```python
df.describe()
```

```python
df.dtypes
```

```python
## Handling Missing Values
df.isnull().any()
```

```python
df.isnull().sum() ## No of missing values
```

```python
df_filled=df.fillna(0)
```

```python
## Filling values with the mean of the column
df['Sales_fillNA']=df['Sales'].fillna(df['Sales'].mean())
```

```python
df.dtypes
```

```python
## Renaming Columns

df=df.rename(columns={'Date':'Sale Date'})
df.head()
```

```python
## Cange datatypes

df['Value_new']=df['Value'].fillna(df['Value'].mean()).astype(int)
df.head()
```

```python
df['New Value']=df['Value'].apply(lambda x:x*2)
df.head()
```

```python
## Data Aggregating and grouping 

df.head()
```

```python
grouped_mean=df.groupby('Product')['Value'].mean()
print(grouped_mean)
```

```python
grouped_sum=df.groupby(['Product','Region'])['Value'].sum()
print(grouped_sum)
```
