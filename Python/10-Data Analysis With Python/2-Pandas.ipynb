### Pandas-DataFrame And Series

Pandas is a powerful data manipulation liabrary in Python , widely used for data analysis and data cleaning . It provides two primary data structures: Series and DataFrame. A Series is a one dimensional array-like object , while a DataFrame is a two dimensional, size-mutable and potentially heterogenous tabular data structure
with labeled axes (rows and columns).

```python
## Series
## A Pandas Series is a one dimensional array-Like object that can hold any data type. It is similar to a columnn in a Table.

import pandas as pd

data=[1,2,3,4,5]
series=pd.Series(data)
print("Series: \n " , series)
print(type(series))
```

```python
## Create a series from Dictionary
data={'a':1,'b':2,'c':3}
series_dict=pd.Series(data)
print(series_dict)
```

```python
data=[10,20,30]
index=['a','b','c']
pd.Series(data,index=index)
```

```python
## DataFrame
## Create a dataframe from a dictionary of list

data={
    'Name':['Raj','Sharma','John'],
    'Age':[22,33,44],
    'City':['Banglore','NewYork','Florida']
}

df=pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np
np.array(df)
```

```python
## Create a DataFrame from a list of Dictionaries

data=[
    {'Name':'Raj','Age':32,'City':'Banglore'},
    {'Name':'Sharma','Age':32,'City':'Banglore'},
    {'Name':'Krish','Age':32,'City':'Banglore'},
    {'Name':'Jack','Age':32,'City':'Banglore'},
]

df=pd.DataFrame(data)
print(df)
print(type(df))
```
