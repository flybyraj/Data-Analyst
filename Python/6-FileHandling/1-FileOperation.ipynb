### File Operation- Read and Write Files

File handling is a crucial part of any programming language. Python provides built in function and methods to read from and write to files, both text and binary. This lesson will cover the basics of file handling , including reading and writing text files and binary files.

```python
## Read a whole file

with open('example.txt','r') as file:
    content=file.read()
    print(content)
```

```python
## Readd a file line by line

with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) ## strip() removes the newline character
```

```python
## writing a file(overwriting)

with open('example.txt','w') as file:
    file.write('Hellow World\n')
    file.write('this is a new line.')
```

```python
## Write a file without overwriting

with open('example.txt','a') as file:
    file.write("\n Append operation taking place")
```

```python
## Writing a list of lines to a file

lines=['First Line \n','Second Line\n' , 'Third Line\n']
with open('example.txt','a') as file:
    file.writelines(lines)
```

```python
## Binary Files

data=b'\x00\x01\x02\x03\x04'
with open('example.bin','wb') as file:
    file.write(data)
```

```python
## Read a content from source text file and write to a destination text file

# Copying a text file
with open('example.txt','r') as source_file:
    content=source_file.read()

with open('destination.txt','w') as destination_file:
    destination_file.write(content)
```
