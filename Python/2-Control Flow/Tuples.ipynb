Introduction to Tuples

Explanation:

Tuples are ordered collection of items that are immutable . They are very similar to lists , but their immutability makes them different.

```python
## creating a tuple

empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst=list()
print(type(lst))
tpl=tuple()
print(type(tpl))
```

```python
numbers=tuple([1,2,3,4,5,6])
print(numbers)
print(type(numbers))
```

```python
list((1,2,3,4,5,6))
```

```python
mixed_tuple=(1,"Hello World",3.14,True)
print(mixed_tuple)
print(type(mixed_tuple))
```

```python
## Accessing Tuple Elements

numbers
```

```python
print(numbers[2])
print(numbers[-1])
```

```python
numbers[0:4]
```

```python
numbers[::-1]
```

```python
## Tuples Operation

print(numbers)
print(mixed_tuple)
```

```python
concatination_tuple= numbers + mixed_tuple
print(concatination_tuple)
```

```python
mixed_tuple * 3 
```

```python
numbers * 3
```

```python
## Immutable nature of Tuples
## Tuples are immutable , meaning their elements cant be changed once assigned.

lst=[1,2,3,4,5]
print(lst)

lst[1]="Raj"
print(lst)
```

```python
print(numbers)
print(type(numbers))
numbers[1]="Raj"
print(numbers)
```

```python
## Tuple Methods

print(numbers.count(1)) ##Returns no of occurence of value
print(numbers.index(3))  ##Returns first index of value
```

```python
## Packing and unpacking Tuple
## Packing

packed_tuple=1,"Hello",3.14
print(packed_tuple)
```

```python
## Unpacking a Tuple
a,b,c=packed_tuple

print(a)
print(b)
print(c)
```

```python
## Unpacking with *
numbers=(1,2,3,4,5,6)
first,*middle,last=numbers
print(first)
print(middle)
print(last)
```

```python
## Nested Tuple

lst=[[1,2,3,4],[6,7,8,9],[1,"Hello",3.14,"c"]]
lst[0][0:3]
```

```python

lst=[[1,2,3,4],[6,7,8,9],(1,"Hello",3.14,"c")] ## Tuple inside List
lst[2][0:3]
```

```python
## Tuple inside Tuple

nested_tuple=((1,2,3),("a","b","c"),(True,False))

## Access the elements inside a tuple
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## Iterating over nested tuples
for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```

#### Conclusion

Tuples are versatile and useful in many real_world scenarios where an immutable and ordered collection of items is required . 
They are comonly used in data structures , functions arguments and return values , and as dictonary keys .
Understanding how to Leverage tuples effectively can improve the efficiency and readability of your Python Code.
