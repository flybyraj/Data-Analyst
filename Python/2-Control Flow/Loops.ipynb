```python
##For Loop
##loop used to iterate over a sequence of numbers

range(5)
```

```python
for i in range(1,6):
    print(i)
```

```python
for i in range(1,10,2):
    print(i)
```

```python
for i in range(10,1,-2):
    print(i)
```

```python
##Strings

str="Raj Sharma"

for i in str:
    print(i)
```

```python
##While Loop
##continues to execute as long as the condition is True.

count=0
while count<5:
    print(count)
    count=count+1
```

```python
count=0

while count%2==0:
    print(count)
    count=count+1
```

```python
##Loop control Statements

##Break
##the break statement exits the loop permanently
for i in range(10):
    if i==5:
        break
    print(i)
```

```python
##continue
##skips the current iteration and continues with the next.

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
##Pass
##The pass statement is a nulll operation;it does nothing

for i in range(5):
    if i==3:
        print("The number is", i)
        pass
    print(i)
```

```python
##Nested Loop
##Loop inside a Loop

for i in range(3):
    for j in range(2):
        print(f"i: {i} and j:{j}")
```

```python
##Examples - calculate the  sum of first N natural numbers usiing a while and for loop
##While Loop

num=int(input("Enter the NUmber: "))
count=1
sum=0

while count<=num:
    sum=sum+count
    count=count+1

print(f"Sum of first natural numbers upto {num} is {sum}")
```

```python
##For Loop

n=int(input("Enter the NUmber: "))
sum=0

for i in range(n+1):
    sum=sum+i

print(f"The sum of First Natural Numbers upto {n} is {sum}")
```

```python
## Prime numbers bw 1 and 100
for num in range (1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)
```
