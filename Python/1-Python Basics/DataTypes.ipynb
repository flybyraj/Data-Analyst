DATA TYPES
Defination - Data types are classification of data which tells the compiler or interpreter how the program intends to use data.

It determines the type of operation that can be perforemd i=on the data, the value the data can take and amount of memory needed to store the data.

```python
## Integer Example

age=35
type(age)
```

```python
## Floating Point DataType
height=5.11
print(height)
print(type(height))
```

```python
## String DatType Example

name="Raj"
print(name)
print(type(name))
```

```python
## Boolean DataTypes
is_true=True
type(is_true)
```

```python
a=10
b=10

type(a==b)
```
